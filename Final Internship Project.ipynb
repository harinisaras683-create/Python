```python
df['Salary_Group'] = pd.cut(
    df['Monthly_Salary'],
    bins=[0,5000,10000,50000],
    labels=['Low Salary','Medium Salary','High Salary']
)

pd.crosstab(df['Salary_Group'], df['Attrition'])
```

```python
import matplotlib.pyplot as plt

df['Attrition'].value_counts().plot(
    kind='pie',
    autopct='%1.1f%%'
)

plt.title('Attrition Yes vs No')
plt.ylabel('')
plt.show()
```

```python
import matplotlib.pyplot as plt

df['Attrition'].value_counts().plot(
    kind='pie',
    autopct='%1.1f%%',
    figsize=(6,6)
)
plt.title('Attrition Rate')
plt.ylabel('')
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

dept_attrition = pd.crosstab(df['Department'], df['Attrition'])

dept_attrition.plot(kind='bar', figsize=(8,5))
plt.title('Department-wise Attrition')
plt.xlabel('Department')
plt.ylabel('Count')
plt.show()
```

```python
df['Age_Group'] = pd.cut(
    df['Age'],
    bins=[20,30,40,50,100],
    labels=['20-30','31-40','41-50','51+']
)

age_attrition = pd.crosstab(df['Age_Group'], df['Attrition'])

age_attrition.plot(kind='bar', figsize=(8,5))
plt.title('Age Group Analysis')
plt.xlabel('Age Group')
plt.ylabel('Count')
plt.show()
```
