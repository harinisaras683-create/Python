```python
attrition_rate = (df["Attrition"] == "Yes").mean() * 100

print(f"Overall Attrition Rate: {attrition_rate:.2f}%")
```

```python
dept_attrition = (
    df.groupby("Department")["Attrition"]
    .apply(lambda x: (x == "Yes").mean() * 100)
    .sort_values(ascending=False)
)

print(dept_attrition)
```

```python
salary_bins = [0, 30000, 70000, float("inf")]
salary_labels = ["Low Salary", "Medium Salary", "High Salary"]

df["Salary_Group"] = pd.cut(
    df["Monthly_Salary"],
    bins=salary_bins,
    labels=salary_labels
)

salary_attrition = (
    df.groupby("Salary_Group")["Attrition"]
    .apply(lambda x: (x == "Yes").mean() * 100)
)

print(salary_attrition)
```

```python
job_satisfaction = (
    df.groupby("Job_Satisfaction")["Attrition"]
    .apply(lambda x: (x == "Yes").mean() * 100)
)

print(job_satisfaction)
```

```python
overtime_attrition = (
    df.groupby("Overtime")["Attrition"]
    .apply(lambda x: (x == "Yes").mean() * 100)
)

print(overtime_attrition)
```

```python
import matplotlib.pyplot as plt

df["Attrition"].value_counts().plot(
    kind="pie",
    autopct="%1.1f%%",
    figsize=(6, 6),
    ylabel=""
)

plt.title("Attrition: Yes vs No")
plt.show()
```

```python
department_attrition = (
    df.groupby("Department")["Attrition"]
    .apply(lambda x: (x == "Yes").mean() * 100)
).sort_values(ascending=False)

department_attrition.plot(
    kind="bar",
    figsize=(8, 5)
)

plt.title("Attrition by Department")
plt.xlabel("Department")
plt.ylabel("Attrition Rate (%)")
plt.xticks(rotation=45)

plt.show()
```

```python
salary_attrition = (
    df.groupby("Salary_Group")["Attrition"]
    .apply(lambda x: (x == "Yes").mean() * 100)
)

salary_attrition.plot(
    kind="bar",
    figsize=(7, 5)
)

plt.title("Attrition by Salary Range")
plt.xlabel("Salary Group")
plt.ylabel("Attrition Rate (%)")

plt.show()
```

```python
plt.figure(figsize=(8, 5))

plt.scatter(
    df["Total_Work_Experience"],
    df["Monthly_Salary"],
    alpha=0.5
)

plt.title("Experience vs Salary")
plt.xlabel("Experience (Years)")
plt.ylabel("Monthly Salary")

plt.show()
```
